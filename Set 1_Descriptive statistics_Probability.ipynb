## 5. Returns on a certain business venture, to the nearest $1,000, are known to follow the following probability distribution

|x|	|P(x)|
|:-----------------------------|	|:-----------------------------|
|-2000|	|0.1|
|-1000|	|0.1|
|0|	|0.2|
|1000|	|0.2|
|2000|	|0.3|
|3000|	|0.1|




```python
x1=[-2000,-1000,0,1000,2000,3000]
p_x1=[0.1,0.1,0.2,0.2,0.3,0.1]
```

## (i)	What is the most likely monetary outcome of the business venture?
- The returns with highest probability is the likely monetary outcome of the venture so it is 2000
## (ii)	Is the venture likely to be successful? Explain
- The probability of getting returns without loss from the venture is P(x>0)=0.8,0.8-(.1+.1)=0.6 , which is quite a higher percentage of profit or no-loss so the venture is likely to be successful
## (iii)	What is the long-term average earning of business ventures of this kind? Explain
- E(x)=sum(P(x) * X)


```python
sum1=0
for i in range(len(x1)):
    sum1=sum1+(x1[i]*p_x1[i])
sum1
```

## (iv)	What is the good measure of the risk involved in a venture of this kind? Compute this measure
- The good measure of the risk involved in a venture of this kind is measure by Variance and Standard Deviation

```python
sum2=0
for i in range(len(x1)):
    sum2=sum2+(x1[i]*x1[i]*p_x1[i])
sum2
```

```python
# Var(x)=E(x^2)-E(x)^2
var1=sum2-sum1**2
var1
```

```python
sd1=np.sqrt(var1)
sd1
```
